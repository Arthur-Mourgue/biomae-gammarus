```python
import os
import glob

# CONFIG
LABEL_DIR = "./ds_gammare/val/labels" # Chemin vers tes .txt
NEW_CLASS_ID = "0" # On force tout le monde à devenir la classe 0

def merge_labels():
    files = glob.glob(os.path.join(LABEL_DIR, "*.txt"))
    count = 0
    
    for file_path in files:
        with open(file_path, "r") as f:
            lines = f.readlines()
            
        new_lines = []
        for line in lines:
            parts = line.strip().split()
            if len(parts) > 0:
                # On remplace l'ID (premier élément) par 0
                parts[0] = NEW_CLASS_ID
                new_line = " ".join(parts) + "\n"
                new_lines.append(new_line)
        
        # On écrase le fichier avec les nouvelles lignes
        with open(file_path, "w") as f:
            f.writelines(new_lines)
            count += 1

    print(f"✅ Terminé ! {count} fichiers convertis en classe unique 'gammare'.")

if __name__ == "__main__":
    merge_labels()
```
